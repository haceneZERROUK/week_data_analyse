```python
# la part des types de paiement dans les paiements

moyen_paiement = Order_payments.groupby('payment_type')['payment_type'].count()

# moyen_paiement

data = moyen_paiement
colors = sns.color_palette("pastel")
plt.pie(data, labels=moyen_paiement.index, colors=colors,autopct='%1.1f%%')

# Ajouter une lÃ©gende
plt.legend(title='Moyen de Paiement', bbox_to_anchor=(2, 0.5))

# Afficher le graphique
plt.show()
```
