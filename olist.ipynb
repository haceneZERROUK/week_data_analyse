```python
#  Proportion des livraisons en retard ou à l'heure.

df_livraison_temps = {
    'correct': (( Orders['order_estimated_delivery_date']-Orders['order_purchase_timestamp'])-(Orders['order_delivered_customer_date']-Orders['order_purchase_timestamp'])).dt.days>= 0,
}

df_livraison_temps = pd.DataFrame(df_livraison_temps)

# Compter le nombre de livraisons en retard
Livraison_correct = df_livraison_temps['correct'].sum()
Livraison_correct_pourcentage = round((Livraison_correct/len(df_livraison_temps)) * 100, 2)
livraison_retard = len(df_livraison_temps) - Livraison_correct
livraison_retard_pourcentage = round((livraison_retard/len(df_livraison_temps)) * 100, 2)

correct = Livraison_correct_pourcentage
retard = livraison_retard_pourcentage
part_livraison = [
[correct] , [retard]
]
part_livraison = pd.DataFrame(part_livraison)
# Renommer la colonne 'pourcentage'
part_livraison = part_livraison.rename(columns={0: 'pourcentage'})

# Renommer les index
part_livraison = part_livraison.rename(index={0: 'correct', 1: 'retard'})

# part_livraison

data = part_livraison['pourcentage']

# Créer un camembert avec Seaborn
plt.figure(figsize=(8, 8))
sns.set(style="whitegrid")  # Utiliser un fond blanc pour le graphique

# Utiliser pieplot de Seaborn
plt.figure(figsize=(4, 4))
colors = sns.color_palette("pastel")
plt.pie(data, labels=part_livraison.index, colors=colors, autopct='%1.1f%%', startangle=90)

# Ajouter une légende
plt.legend(title='part_livraison', bbox_to_anchor=(2, 0.5))

```

```python
# Temps de livraison par région.
livraison_par_region = Customers.merge(Orders, on='customer_id', how='inner').merge(Order_items, on = 'order_id', how = "inner").merge(Product,on = 'product_id', how = 'inner')
livraison_par_region['order_delivered_customer_date'] = pd.to_datetime(livraison_par_region['order_delivered_customer_date'])
livraison_par_region['order_purchase_timestamp'] = pd.to_datetime(livraison_par_region['order_purchase_timestamp'])

livraison_par_region['delai_livraison'] = (livraison_par_region['order_delivered_customer_date'] - livraison_par_region['order_purchase_timestamp']).dt.days

# Groupement par région et calcul de la moyenne des délais de livraison
# livraison_par_region = livraison_par_region.groupby('customer_state')['delai_livraison'].mean()
# livraison_par_region['delai_livraison'] = round(livraison_par_region['delai_livraison'], 2)
moyenne_par_region = livraison_par_region.groupby('customer_state')['delai_livraison'].mean()
moyenne_par_region = moyenne_par_region.round(0)
moyenne_par_region = pd.DataFrame(moyenne_par_region)
moyenne_par_region = moyenne_par_region.sort_values(by='delai_livraison')
# moyenne_par_region
plt.figure(figsize= (20,5))
sns.barplot(x= 'customer_state', y = 'delai_livraison', data = moyenne_par_region, palette= 'magma', legend= True)
```
